Now we will drop unnecessary columns - Title and ID

```python
df.columns = ['ID', 'Title', 'Genre', 'Description']
df.head()
```

```python
df.isna().sum()
df.dropna(inplace = True)
print(df.shape)
```

**Clean the description as it is our source**
1. Convert it to lower case
2. Remove punctuations
3. Tokenize
4. Remove Stopwords
5. Apply lemmatization

```python
#importing nltk dependencies
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
# function to do the above stuff
import string
def processor(text):
    # Convert to lower case
    text = text.lower()
    # Divide the text into tokens
    tokens = word_tokenize(text)
    # Get all stop words and remove them
    stop_words = set(stopwords.words('english'))
    # Perform lemmatization to reduce the words to their root forms
    lemmatizer = WordNetLemmatizer()
    processed_tokens = [lemmatizer.lemmatize(word) for word in tokens if word not in stop_words]
    processed_text = ' '.join(processed_tokens)
    # Remove punctuation and add space after removing
    cleaned_text = ''.join([char if char not in string.punctuation else ' ' for char in processed_text])
    return cleaned_text
```

```python
#Apply the above processor function to every row of description column in out dataframe
df['Processed_Description'] = df["Description"].apply(processor)
```

```python
df.head()
```

```python
#Check if the newly added column has any null values
df.isna().sum()
```

**Feature Extraction**

```python
df = df.drop(['ID', 'Title'], axis = 1)
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df['genre_encoded'] = label_encoder.fit_transform(df['Genre'])
```

```python
df.head()
```

**Splitting the data and Model training**

```python
from sklearn.model_selection import train_test_split
X = df['Description']
Y = df['genre_encoded']

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2)
```

```python
#Use Tfidf vectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer(max_features=5000)

x_train = tfidf_vectorizer.fit_transform(x_train)
x_test = tfidf_vectorizer.transform(x_test)
```

Using Logistic Regression as the classification model

```python
from sklearn.linear_model import LogisticRegressionCV
Cs = [0.1, 1, 10]
lrcv = LogisticRegressionCV( max_iter = 1000, Cs=Cs, cv=5)
lrcv.fit(x_train, y_train)
```

Evaluating the model

```python
y_pred = lrcv.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
accuracy = accuracy_score(y_test, y_pred)

print(accuracy)
```

**Testing the model**

```python
new_description = '''In the isolated town of Ravenswood, nestled deep in the woods, a dark secret haunts the residents.
  For centuries, eerie whispers have spoken of a malevolent force that awakens every century, seeking vengeance.
  When a group of unsuspecting strangers stumbles upon the town, they soon realize they are trapped in a nightmarish web of terror.
  As shadows lengthen and the unknown draws nearer, they must confront their deepest fears to uncover the horrifying truth that binds them to Ravenswood.
  With every creak and rustle, the boundary between the living and the dead blurs, and survival becomes a race against the malevolent force that hungers for their souls.'''

#process the text
processed_descrption = processor(new_description)

#apply tfidf vectorization to the processed text
test_ready_desc = tfidf_vectorizer.transform([processed_descrption])

#Predict the genre using the trained model
prediction = lrcv.predict(test_ready_desc)

#Convert the predicted label back to the orginal name using inverse_transform
predicted_genre_label = label_encoder.inverse_transform(prediction)[0]
```
